```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = pd.read_csv("tips.csv")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"], kde=False, bins=30)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
#histogram of x var on top, histogram of y var on side, and scatterplot of both in center
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") #kind arg changes central chart type to hexagonal
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg") #linear regression
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

```python
fig = sns.pairplot(tips, hue="sex", palette="coolwarm") #will do different types of charts against all the numerical data it finds
```

```python
fig.savefig("cool_charts.jpg")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
tips.head() #focus on categorical columns
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) #average is the default but can change the estimator object w/ additional arg
```

```python

sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) #estimator is explicitly counting the number of ocurrences, so only need to state x-val
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips) #similar to boxplot, but see the kde >> not as intuitive!
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue="sex", split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color="black") #not great for large data sets
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind="bar") #more general plotting method
```

```python
flights.head()
```

```python
#for a heat map to work, data needs to be in a matrix form
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

```python
fp = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap="magma")
```

```python
sns.heatmap(fp, cmap="magma", lw=.5, linecolor="white")
```

```python
sns.heatmap(fp, cmap="coolwarm", lw=1.5, linecolor="white")
```

```python
sns.clustermap(fp) #re-ordered based on similarity
```

```python
sns.clustermap(fp, cmap="coolwarm", standard_scale=1)
```

```python
#Grids
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
#more control than pairplot
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill') #these 2 lines need to be done in the same cell for it work
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
